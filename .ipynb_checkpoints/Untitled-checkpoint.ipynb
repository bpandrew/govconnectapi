```python
import requests
import json
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize
import datetime
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

## Download and create a dataset of contracts

```python
#session = {"endpoint":"https://governmentconnect.herokuapp.com/"}
session = {"endpoint":"http://127.0.0.1:5000/"}#

QUERY_ENDPOINT = session['endpoint']+"contracts_data?page=0"
uResponse = requests.get(url = QUERY_ENDPOINT)
query_data = json.loads(uResponse.text)
pages = int(query_data['pages'])
pages = 100 # force the maximum number of pages ***** TEMP ******

df = pd.DataFrame()

for i in range(pages):
    QUERY_ENDPOINT = session['endpoint']+"contracts_data?page="+str(i+1)
    uResponse = requests.get(url = QUERY_ENDPOINT)
    query_data = json.loads(uResponse.text)
    df_temp = json_normalize(query_data['data'])
    df = pd.concat([df, df_temp], sort=False)

df = df.drop(['branch', 'division', 'son', 'unspsc'], axis=1) # drop the json 'column headers'
df['contract_value'] = df['contract_value'].astype(float)
df
```

## Find the average cost of contracts in this category

```python
agency_id = 35
category_id = 144

df_op = df.copy()
df_op = df_op[df_op['agency.id']==agency_id]
df_op = df_op[df_op['unspsc.id']==category_id]
df_op
```

```python
avg_value = df_op['contract_value'].mean()
avg_value
```

```python
df_temp = df_op.groupby(['division.id', 'division.title', 'branch.id', 'branch.title']).mean()
df_temp
```
